```python
import pandas as pd
import sklearn
```

```python
df=pd.read_csv("/content/Compression_ML_Dataset.csv",header=1)
```

```python
df
```

```python
df.columns
```

```python
df=df.drop(["Unnamed: 0",'Unnamed: 9'],axis="columns")
```

```python
df
```

```python
df=df.dropna( axis=0, how="any", thresh=None, subset=None, inplace=False)
```

```python
dummies1 = pd.get_dummies(df)
dummies1
```

```python
final=dummies1.drop(['Infill Pattern _Concentric'],axis="columns")
```

```python
final
```

```python
x= final.drop(['Failure Force','Density','Thermal Cond'],axis='columns')
x
```

```python
y=final['Failure Force']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```
